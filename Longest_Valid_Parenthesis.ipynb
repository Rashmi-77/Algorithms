# Longest valid Parentheses
Given a string S consisting of opening and closing parenthesis '(' and ')'. Find length of the longest valid parenthesis substring.

```python
def longest_valid_parenthesis(S):
    
    N = len(S)
    open_p = 0
    close_p = 0
    pairs = 0

    for i in range(N):
        if S[i] == '(':
            open_p+=1
        else:
            close_p+=1

        if open_p == close_p:
            pairs = max(pairs,open_p*2)
        elif open_p < close_p:
            open_p  = 0
            close_p = 0

    open_p = 0
    close_p = 0

    for i in range(N-1,-1,-1):
        if S[i] == ')':
            close_p+=1
        else:
            open_p+=1

        if open_p == close_p:
            pairs = max(pairs,close_p*2)
        elif close_p < open_p:
            open_p  = 0
            close_p = 0
    
    return(pairs)
        
S = '((()'
print(longest_valid_parenthesis(S))

S = ')()())'
print(longest_valid_parenthesis(S))

S = ')(((())(()()(()))))()(()))(((()))((()()(((()())()()())()))((()()()((()()()())()()))))(()())))())))(()))((()()))())((()()(((((((((()(()()(())()((()))()()()()(((()))()()))))()((((())(()(()()((((())(()())((()))))))((())()()()())(())()))((())())(())))()((())()))))()))))()))))(())(((()())(((()((())()))))))))(()()()))(()))()(()((((()(()(()())))))()())))))(()()))())))((())())((())())()(((()(())()))))()))())))))()((()))())()))(()((((())())()(()))()))()))))(((())()(()((()()(()(()((((())(((())(()()))(())())((((()()(())))())())))))))))())()()(((()))()))()(()(()(())))))())((((()((()((((()()(()())())(((())((((())((())))((((()(((((())))(()))(()))))()))((()))(((()((())))((()(()((()()()()))()())()(())((()()()))()(((((()(())()(())))(((((()(())(()((()()(()(())())()))(()))())()))(())()()())()))())))())(()())))(((((()()(((((())()))))())()((())))(())((())()((((())(()()))((()((()()((()()))(())()(()(()(())((())()())(()())()))))()(())))(())()()(()(())()((()(((())()))(())()((((())((()))()()))))()(())()((()()(()()(((()())()))))(()(()())(()((()()())())((((((())))()()))((())()))))()()))(((()(()(((()()(((()(((()))(()()()((()()))()))((()()()(((())(()(())())())())((()(()((())(()))(()()(())))()())))(())()((())()(())(((()((())))()()()(((()(((())))((()(()(()(((())))((()()(()(()(((((()()()()())(((())()(((())()()(((()()))))()()()))))((()()(((())((()((((())()(()((())))))(()()))(()((())()(()()()()(()())))())())(((((())())))()((()))))()((()((())))(())())((()())(((()()())((((()((((()()(((()((()))))))))()()(()))())()()())))(()(()(((()()())(())(((()()(()()(((()))((()))))(()()((())(())))(())()())))))()()))()())))())()())))(()((((()())()(())()))))()()(())))((())(()))(()((((()(((((())()()()(((()((())(()))()(()()(()()((())(((()))((()()()))()(()()((())()())(()()(())(((()))())))(((()((((()))))((()))(()(()(()())(())((()))))))(()()())))))((()()(())))())())(()()))()(((())))(())())(((((())())())))((())(()))))(((()))(()))))))))())))))())(((()))))))(())()))))())(()()()))(()(()(()))))))))((((())))(()((()()((())(())(()()()(())()()))(())((()(())))(())()))(())(())(((((())())())()())))()()((()(())((())(()(()())))()))(()))()(()())))))()())((((())))(()()((())())())())()))(((((((())(()(()((()())())((((()()()))))(()()(())))(())((()))()(())()))(()()())(()))))(((((())(()(()(())(())()())()((())))()))))(())))(((((())()()()())))()))))(())((()))(()((()())))))))())()(((()))()())()()))(()((()())(((()))()()))))()()((()))(()((()))(())))())()()())))()))(()))(()()(()(((()))((((((((((())(()()())))()()(((())))((())((((())))())()(())((())(()))((())))())()())))))))()(()))))((()))))())()(()(()())))())()))((((((((((()()())))()()))))))))))(((((()))))(()(()))())))()(()((()()())))))(())()(()(())))()))(()))))))(())))(())))())())))))))())(()()(())((((())())))((((((()((())))(((())())()(()((())()())())()))()(()())(()))()()))()()())())()((())))))(()())()(()()))()(()((())))))()))(()()))()))))())))(()(())()())(()())(())())(((((())()(()))()))()()()(())((())((((((())()))(())(()))(()()(()()((())()))((())()((()()(())(())()())((()))(()()))(()))))))(()())()())))))(())()(()(()(()()())()(((((()((()()))(()))))))()(()))((((((()))())()())))))))(((((((()))((())(())))())())(()()()()())())(()((((((()(()((()())))))())(())))((()))((()((()()())())()))(())((((()()(()))()(((((((())()())()((()()))())(()()))(())()((()(()))()(()(()()((((()))))()()((()(())(()()(((()((()((((()((()))())()(()((()((())))()()()()(((()(()((()))()()(()))()(((())))()))()()(())(((()())(())())())(())())))))()()())))(()()(())()))()(())())))())(()))))()(()())(((())))))))((((()(()(((())((()()())))()(())((()()(())(()((()()))(()()))(()))()))(()()(())()((())))))()()()))))))))())()))()()))))))()(((()))))))(((()(())())()))))())()(()(()()((()()))(()(())()))))())))()(()))(((((())()(())()()())((()()(()()((()))()))()(())(()()()()))(()()((((()(()((()(((())((())((()((()))(()()))()))((((((()(())))((((((()())))()))(((()))()(()(()((()))())())())((()))()((()(((())()(((()(())((())(())(((((()())(())()((()((()))(((())((())((((())(()()(())())((()(())())(())(((())((()(())()(((()(())()(()))((()()()))())()()())))))()))()))(())()(())(((()((((((())))(()(((()()))))((()()))((()(()((()((())())()))()()(()()())))((()(((((()()))))()((((()((()(((()()(((())))))())(((()(())))((()))((())))())(()(()()))())))((()()(((((()))))((()())()))))))())()()()(())(()))((()())()((((((()())))(((())(()()(((((()))((())(()((((())())))()()))(())((((())))))((((()(((()())(())())((()((()((())()))(())()))()(()())))((((()(())(()())(((()))(())())))((())))))((())((()(((())()))))))(()()))))(()(()((()))(((((()(())())(()(((((()()())))(()()(((((()((()()(((()())(()(())(()()()())))(((())(())))((())())))()((()(()))((())))()()()()(()()((())((((()()((((())))()((((()()())()((()()()()(()(()()))(())))()(())))(())()))())))())()(()()()())()))(())()())()()((((((()()))((((()())((()((())(()()()(((()()())))((()))()())))()(((()))()()(()))(((((()()(()()))(((()(())()(()()()()()))(()))()()(()))()((()(())(()(()())())))((()(())((((())()((())(((((((()()(()()))()(()()((()()(((()))()())()(()())((()))(()((())))(((()(((())))()()(()(()()(())()()()(())))()))(())(())(()))))))()()()))()(((())((()((()()(()(()(()(()())))(()))((()()())))(((())(()())())(((()))(((()))))((()()(())()()()()(()))(()(()((()(()()(())(()))))(((())())))()()())))(()()()(()((())((((()())))())()())())((((())(()))(()())()()())))(((()))((()())()((()(()(())()()()()((()()))))()))()(()())((()()()())))()(()())(()))()(((())())))())(((())((()())()(()))())))((()(())(()(())(()()()(())))))((())(()()()))))((()()))())(((())()))()())())()))()((())()()(()))))))(()(()())(()))(()))((())()))()))))(())()))))()(()()()())()))())()()))()))()((()()()())(((())()(()(())())))((())((()((()))))((()()(((((((((()((()))()(()))((()))((()))()()((())(())())()))((((()())()((()))(()()((((()((())))()(())(())())((()()()((())))())))()))(())())))()((()))()((()(()(()(())()(()(()((((()()))(())((()()(())()()(()()()(((()))()))(()((()(()(()())))()((()(())))(((())((()(((()(()()))(())()(()))(((())((()))))()(())(()()()(())()())((()())()((()(((()((((())(())))(()(()())))()(())(()))()))))))(((())(((((()()((())()()(()())((())((()))(())()))))(((()((((()))))(((((())(()(())()(((()()())(()())))(()))()))))))(())(())()()))())()(()(()(()((())()())(()))()(())((((()())))(()))(()((()((())))))))(()(()(()))))(()))())(()())))))(())((()(()((()()(()())(()(())((()))((()((()(()()((((())())()((((()))()()()(())(()()(()(())()((()()()))))))()))()(()()())())(()()))))())()()()((((()()(()()()))))())))((()())(())()((()()))))())))()()())()()(())(()(((())))()((()()))))()((()())))()())()))((()())(())))()()(())(()(())))(())((((()(()())))))))(())()(()))))()()((())))))())())())))(((((())(())(()((((()))()))())()()())()(()))()()())())(()()))()())))()()(()))((()(()()((((((()(()(((()))))()))()()))))()()()))()())(()(())())))(((((()()((((())(())))(()))())()()((()))((()(()))))))((((()))()()()(((()))(()(())(()()())())()(()))(())()(()))()))(())())))()))))))()))((((()))))(((()(()))())()(((((())())())((())))()(())(((()()()))(()())((()(()(()))())))()()(()()(((()))()((())(())())()))))))()(())()))((()((())))))()())())))(()()((())))))())())))(()()))))()(()))()())()(()())()())())())))(((())(()())(())(((()((((()()()()(()()())()(((()))))(())()(((()))())))))()()()(()()())(()))))())))))))))(())(()()))()()))())))))((((((()()()(())()(()(()))))))((()((()()))()(())(((()((((((())(()))))))()(())())))()((())(()())))))()))()(())())()))())(())))))()))(())((((())())()((())()(()()()))()))((()(((((()))((()()()()(()()(())))((()((((()))())(())(())))(((((()))((())())(()())((()(()()()))()()()()())()))())(((()())(()()))()(((()))))))())))())))))(((((()(()))))))))))())())()))))()()()())(())(())()()())(())(())(())((((())((((())))())))()()))())(())(()()))(()()())(())())))())(()()()(()((())((())))(()(()((()((((((()()(()))(())((()()))(((())()()())((()((())((()))()))(()))()((((()()))))()()())(())()(()())()()((()))))()(((())))))))()())))(((()(()(((()())(()(())())()(()))(((()()(()))))(())(())(((()(()()))())))())())))))(((())(())))(()))())(()()(()(())((((()(()()))))(()()()((()())()((()))())())()))))))()))()(()((()))))()())()())())())()())))(()())()()())()(())(()((()(()()(())()(((((())((()))()))()(()((()(((())))())((()((()((((())((()()()((((()(()(()))(()))(()())))()()((())())((()((()()))((()(())()))((((((())))()))()))(((()()())(((()))())()())())(()))()()))()()())()((()))))(())))))))())(()(((((()))())((())())))))())(()))(()())((()()())())())()(()(()(())(((())))()()))))()(())))((()((()))))))((()(((())((((())(((()))())())(()))()(()()(())()()((((()()(())()(()()(((())()(()))((()())))()((()()))(()()()(()((()(((()((()))()))))())())((((())(((((())))(()(())(()))))((())())()()(()))(()())((((()((()))())(())()(()(((()((()())((()())))((((()(((()))))((())()()())())(())(()))))())(())))())))())((())((()()(())))(()(())()())()()((()))()))())((()()))(())()()()((()(((()(()()()(()((()(())()((())()()((()))(()))((()(()(((())(())((()()((())()()()()(()))(())))((()()))(())()(((())(()((()())((((((((((()()))()()))())(())(()))((()((()())()(()(()(())(())))()()))(((())(()(()))(()((())()))()(()))())((())))((()(())()()))()())())()((((()))((())())(()((())())()(()))())()((((()(((((())((((()))()()(()(())((((((())))((()))()()))))((()((())((()()))((((()))(()(()))()()))((()(()((())(()((())))))))())()()()()))()))(())()(()((((()()(((((()())())((((()()()(((()()))))(((()((((()))))(())()()()(()(())))((())(((()((())())()(())(()))))((()))(()()())())((())((()(()(()(()((()(()(())(()))))(()()())()()))(()())))))(())(()))))())(())()))()())()))())((('
print(longest_valid_parenthesis(S))

S = '()((()()))'
print(longest_valid_parenthesis(S))

S = '))())(('
print(longest_valid_parenthesis(S))
        
```
